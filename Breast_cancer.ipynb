```python

_, ax = plt.subplots()
scatter = ax.scatter(bc.data[:, 0], bc.data[:, 1], c=bc.target)
ax.set(xlabel=bc.feature_names[0], ylabel=bc.feature_names[1])
_ = ax.legend(
    scatter.legend_elements()[0], bc.target_names, loc="lower right", title="Classes"
)
```

```python
print (type(bc.data))
print (type(bc.target))
```

```python
X=bc.data
y=bc.target
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
print(knn)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=4)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn import metrics
knn.fit(X_train, y_train)
y_preds = knn.predict(X_test)
print(metrics.accuracy_score(y_test, y_preds))
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(solver='liblinear')
logreg.fit(X_train, y_train)
y_preds = logreg.predict(X_test)
print(metrics.accuracy_score(y_test, y_preds))
```

Logistic regression gives a more accurate prediction than K nearest neighbors classification algorithm

```python
from sklearn.tree import DecisionTreeClassifier
dc=DecisionTreeClassifier()
dr=dc.fit(X_train,y_train)
y_pred = dr.predict(X_test)
print(metrics.accuracy_score(y_test, y_pred))
```
